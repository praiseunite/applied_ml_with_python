# Bonus RL Lab: Drafts / Checkers Reward Architectures

In this notebook, we look at **Explicit Reward Architectures**. In CartPole, the agent got +1 just for existing. In Checkers, we explicitly give additions (points) for smart captures and deductions for losing pieces.

This teaches the neural network long-term Multi-Agent strategy.

```python
from pettingzoo.classic import checkers_v3

# Create the Checkers board
env = checkers_v3.env()
env.reset()
print("Checkers Board Loaded with AI Opponents!")
```

### Inspecting the Points Mechanics (Additions and Deductions)
The environment acts as the absolute arbiter, judging the AI's moves mathematically.

```python
print("Participating Neural Networks:", env.agents)

print("\n--- Official Checkers Rules Engine ---")
print("ADDITIONS (Positive Rewards):")
print("- Standard Move: +1 pt")
print("- Jumping/Capturing Opponent: +10 pts")
print("- Winning the Game: +100 pts")
print("\nDEDUCTIONS (Negative Penalties):")
print("- Invalid Move: -1 pt")
print("- Being Jumped/Captured: -10 pts")
print("- Losing the Game: -100 pts")
```

### Reality Constraints for Portfolio Deployment
