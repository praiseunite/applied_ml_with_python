## 📋 Objectives
1. Train a model in Memory (RAM).
2. **Serialize** it permanently to the Hard Drive.
3. **Deserialize** it back into Memory to prove it retained its intelligence.
4. Understand JSON request architectures.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

print("✅ Libraries Loaded Successfully.")
```

---
## §1 — Training a Basic Model
We need an "Intelligent" object to deploy. Let's do a fast decision tree.

```python
X, y = load_iris(return_X_y=True, as_frame=True)

print("Training Decision Tree...")
my_classifier = DecisionTreeClassifier(max_depth=3, random_state=42)
my_classifier.fit(X, y)

train_acc = accuracy_score(y, my_classifier.predict(X))
print(f"Initial Memory Model Accuracy: {train_acc * 100}%")
```

---
## §2 — Serialization (`joblib` vs `pickle`)

If we close this Jupyter notebook, `my_classifier` is instantly destroyed. We must freeze it.

We use `joblib`. It is significantly faster than standard python `pickle` for Numpy Arrays (which Scikit-Learn uses under the hood).

```python
import joblib
import os

# 1. Save it to disk
filepath = "iris_model.joblib"
joblib.dump(my_classifier, filepath)

# 2. Verify it exists on the hard drive
if os.path.exists(filepath):
    print(f"✅ SUCCESS! Model saved to {filepath}")
    file_size = os.path.getsize(filepath)
    print(f"File Size: {file_size} bytes")
```

---
## §3 — Deserialization (Waking the AI up)

Let's pretend we are in a completely new script. We want to load the model and make a prediction on a new flower.

```python
# Load the object back into RAM
loaded_ai = joblib.load(filepath)
print("✅ AI successfully loaded from disk!")

# Create a new flower (Fake Data)
new_flower = pd.DataFrame([{
    'sepal length (cm)': 5.1,
    'sepal width (cm)': 3.5,
    'petal length (cm)': 1.4,
    'petal width (cm)': 0.2
}])

# Make a prediction WITHOUT having called .fit() in this block!
prediction = loaded_ai.predict(new_flower)
print(f"The AI classified the new flower as class: {prediction[0]}")
```

---
## §4 — Transition to JSON for APIs

In the cell above, we created a Pandas DataFrame manually. But when an iPhone app asks your model for a prediction, iPhones don't know what Pandas is. They send **JSON**.

JSON is identical to Python Dictionaries. A Flask Server simply takes JSON, translates it to Pandas, feeds it to the Model, and translates the answer back to JSON.

```python
import json

# What an iPhone sends over the internet:
json_payload = '{"sepal length": 5.1, "sepal width": 3.5, "petal length": 1.4, "petal width": 0.2}'

# The Flask Server converts JSON text into a Python Dictionary:
python_dict = json.loads(json_payload)
print("Type after parsing:", type(python_dict))

# The Flask Server converts the Dictionary to Pandas:
df_received = pd.DataFrame([python_dict])
display(df_received)
```
