# 🎮 RL Game Agent Lab: CartPole with PPO

In Phase 1 of our Practice Session, we will build the "Brain" of our application. We are going to train an AI agent to perfectly balance a pole on a moving cart using **Reinforcement Learning**.

### What you will learn:
1. How to initialize a Reinforcement Learning environment using `gymnasium`.
2. How to use `stable-baselines3` to train an agent using the PPO (Proximal Policy Optimization) algorithm.
3. How to export your trained agent as a file so it can be deployed into an application.

```python
# Step 1: Install Required Libraries
# We install gymnasium for the environment and stable-baselines3 for the RL algorithms.
!pip install stable-baselines3[extra] gymnasium[classic-control]
```

```python
import gymnasium as gym
from stable_baselines3 import PPO
from stable_baselines3.common.evaluation import evaluate_policy
import os

# Ensure the output directory exists
os.makedirs("../app", exist_ok=True)
print("Libraries imported successfully!")
```

### Step 2: Initialize the Environment
We are using `CartPole-v1`. The goal is to keep the pole upright. The environment gives a reward of +1 for every step the pole stays balanced.

```python
env = gym.make("CartPole-v1")
print("Observation space:", env.observation_space)
print("Action space:", env.action_space) # 0: Push Cart Left, 1: Push Cart Right
```

### Step 3: Train the Agent
We load the `PPO` algorithm. MlpPolicy means it uses a standard Neural Network. We train it for 25,000 timesteps.

```python
# Initialize the RL Algorithm (Brain)
model = PPO("MlpPolicy", env, verbose=1)

# Train for 25,000 timesteps (takes about 15-30 seconds on most computers)
model.learn(total_timesteps=25000)
print("Training completed!")
```

### Step 4: Evaluate the Agent
Let's see how smart our agent became. A perfect CartPole score is 500.

```python
mean_reward, std_reward = evaluate_policy(model, env, n_eval_episodes=10)
print(f"Mean reward: {mean_reward} +/- {std_reward}")
```

### Step 5: Export the "Brain" for Deployment
We save the model so that our Streamlit app can use it in Phase 2.

```python
model_path = "../app/ppo_cartpole"
model.save(model_path)
print(f"Model successfully saved to {model_path}.zip")
env.close()
```
