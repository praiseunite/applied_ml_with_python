---
## 1. Central Tendency & Dispersion Indicator Analysis

Let's load the famous 'Diamonds' dataset. It contains prices and physical attributes of almost 54,000 diamonds. This is perfect for analyzing shape and dispersion.

```python
df = sns.load_dataset('diamonds')
display(df.head())

# THE MOST IMPORTANT FUNCTION IN PYTHON EDA:
display(df.describe())
```

### Interpreting `.describe()`:
1. **Mean vs 50% (Median)**: Look at the `price` column. The Mean is `$3,932`, but the Median (50%) is only `$2,401`. 
2. **What does this mean?** It means the data is **Right-Skewed**. A few incredibly expensive diamonds are pulling the average up, but the "typical" diamond is much cheaper.

---
## 2. Advanced Shape Indicators: Skewness & Kurtosis
Let's mathematically prove our suspicion about the `price` column using Applied Python Functions.

```python
# Filter to numeral columns only
numeric_df = df.select_dtypes(include=[np.number])

print("--- SKEWNESS ---")
# Values > 1 indicate high Right-Skew (Tail extending to the right)
display(numeric_df.skew().sort_values(ascending=False))

print("\n--- KURTOSIS ---")
# Values > 3 indicate heavy tails (lots of outliers)
display(numeric_df.kurtosis().sort_values(ascending=False))
```

```python
# Let's visualize the skewness and Kurtosis of Price!
plt.figure(figsize=(10, 5))
sns.histplot(df['price'], kde=True, bins=50, color='darkred')
plt.title(f"Diamond Prices\nSkew: {df['price'].skew():.2f} | Kurtosis: {df['price'].kurtosis():.2f}")
plt.show()
print("Conclusion: This data is not normally distributed! ML models like Linear Regression would need this data Log-Transformed first.")
```

---
## 3. Advanced Multivariate EDA (`Seaborn` Pairing)
How does the price relate to the Carat (weight) of the diamond? Does the 'Cut' quality change this relationship?

```python
# 1. JointPlot: Relationship + Individual Distributions
sns.jointplot(data=df.sample(2000), x='carat', y='price', hue='cut', palette='viridis', alpha=0.6)
plt.suptitle("JointPlot: Price vs Carat (by Cut)", y=1.02)
plt.show()
```

```python
# 2. FacetGrid: Breaking the plot into panels for deeper clarity
g = sns.FacetGrid(df.sample(2000), col="cut", hue="color", palette="Set1", col_wrap=3)
g.map(sns.scatterplot, "carat", "price", alpha=0.5)
g.add_legend()
plt.subplots_adjust(top=0.9)
g.figure.suptitle("FacetGrid: Relationship split distinctly by Cut Quality")
plt.show()
```
